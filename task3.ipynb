<h4>Bivariate analysis of profit column with the rest of the columns</h4>

```python
# visualization between state and profit column

plt.figure(figsize=(20,10))
df_group=df.groupby(["State"])["Profit"].sum()
df_group.sort_values(ascending=False)
df_group.plot.bar()
plt.show()
```

* <b>As it is clearly seen in the above visualization that states like California and New York has maximum profit whereas Texas and Ohioand has 
least profit among all the states of United States</b>

```python
# visualization between sales and profit column

plt.figure(figsize=(20,10))
sb.scatterplot(df['Sales'],df['Profit'])
plt.show()
```

* <b>Here we observe that after a particular pointincrease in sales dosn't make that impact on profit.
    As you can see above clearly when the sales is at the highest point profit is not that high. </b>

```python
# visualization between ship mode and profit column

plt.figure(figsize=(10,10))
ship_profit=df.groupby(["Ship Mode"])["Profit"].sum()
ship_profit.sort_values(ascending=False)
ship_profit.plot.bar()
plt.show()
```

* <b>Standard class shipping mode is contributing more to increase the profit than other shipping modes</b>

```python
# visualization between city and profit column

plt.figure(figsize=(10,10))
city_profit=df.groupby(["City"])["Profit"].sum()
a=city_profit[city_profit>3000]
a.sort_values(ascending=False)
a.plot.bar()
plt.show()
```

* <b>From above visualisation we conclude that New York city is contributing highest to increase the profit that the rest of the cities </b>

```python
# visualization between Region and profit column

plt.figure(figsize=(10,10))
ship_profit=df.groupby(["Region"])["Profit"].sum()
ship_profit.sort_values(ascending=False)
ship_profit.plot.bar()
plt.show()
```

* <b>By this bar graph we can clearly say that east and west regions contributing more on increasing the profit than rest two regions.</b>
